7. Рассмотрим случайный вектор с формой 100х2, представляющей координаты, и найдем расстояния от точки к точке

````python
def distance_array(points):
    points_1 = points[:-1]
    points_2 = points[1:]

    diff = points_2 - points_1

    squared_diff = diff**2

    distances = np.sqrt(np.sum(squared_diff, axis=1))

    return distances

points = np.random.randint(1, 10, size=(100, 2))


print(distance_array(points))
````

### 3. Визуализация

Используя библиотеки matplotlib/sns/plotly/pandas и набор данных представленный ранее визуализируйте

1. Как менялась средняя цена квартиры в Нагатинском Затоне(Nagatinski Zaton) с течением времени

````python
import matplotlib.pyplot as plt

nagatinski_df = df[df['sub_area'] == 'Nagatinskij Zaton'].copy()
nagatinski_df['timestamp'] = pd.to_datetime(nagatinski_df['timestamp'])
price_by_time = nagatinski_df.groupby(nagatinski_df['timestamp'].dt.to_period('M'))['price_doc'].mean()

plt.figure(figsize=(10,6))
price_by_time.plot()
plt.title('Изменение средней цены квартиры в Nagatinskij Zaton с течением времени')
plt.xlabel('Дата')
plt.ylabel('Средняя цена')
plt.grid(True)
plt.show()
````

2. Постройте гистограмму распределения цен на квартиры с общей площадью более 50 кв.метров

````python
filtered_df = df[(df['full_sq'] > 50)]
data = filtered_df.groupby(filtered_df['full_sq'])['price_doc'].mean()

plt.figure(figsize=(10,6))
plt.bar(data.index, data.values, color='blue', width=50)
plt.title('Зависимость цены от площади квартиры', fontsize=16)
plt.xlabel('Площадь', fontsize=14)
plt.ylabel('Средняя цена', fontsize=14)
plt.grid(True)
plt.show()
````

3. Визуализируйте количество проданных квартир в разных районах

````python
data = df.groupby('sub_area').size()

plt.figure(figsize=(30,8))
plt.bar(data.index, data.values, color='blue', width=1)
plt.xticks(rotation=90, fontsize=12)
plt.xlim(left=-0.5, right=len(data) - 0.5)
plt.title('Количество проданных квартир в разных районах', fontsize=16)
plt.xlabel('Район', fontsize=14)
plt.ylabel('Количество проданных квартир', fontsize=14)
plt.grid(True)
plt.show()
````

4. Постройте box plot отображающий распределение полных площадей квартир в Басманном районе

````python
basmanniy_df = df[df['sub_area'] == 'Basmannoe']

plt.figure(figsize=(8,6))
plt.boxplot(basmanniy_df['full_sq'], patch_artist=True, boxprops=dict(facecolor='blue'))
plt.title('Распределение полных площадей квартир в Басманном районе', fontsize=16)
plt.xlabel('Полная площадь (кв.м)', fontsize=14)
plt.grid(True)
plt.show()
````

5. Визуализируйте зависимость средней стоимости квартир в зависимости от этажа

````python
average_price_per_floor = df.groupby(df['floor'])['price_doc'].mean()

plt.figure(figsize=(10,6))
average_price_per_floor.plot()
plt.title('Средняя стоимость квартиры в зависимости от этажа')
plt.xlabel('Этаж')
plt.ylabel('Средняя цена')
plt.grid(True)
plt.show()
````

7. Постройте график изменения медианной цены на квартиру менее 50 кв. метров за период с 2011 до 2014 года*

\* Задание на доп. балл

````python
filtered_df = df[df['full_sq'] < 50].copy()

filtered_df['timestamp'] = pd.to_datetime(filtered_df['timestamp'])
groups = filtered_df.groupby(filtered_df['timestamp'].dt.to_period('M'))['price_doc'].median()
result_groups = groups[(groups.index <= '2013-12') & (groups.index >= '2011-01')]

plt.figure(figsize=(10,6))
result_groups.plot()
plt.title('Медианная цена на квартиру < 50 кв.м за период с 2011 до 2014')
plt.xlabel('Дата')
plt.ylabel('Медианная цена')
plt.grid(True)
plt.show()
````
